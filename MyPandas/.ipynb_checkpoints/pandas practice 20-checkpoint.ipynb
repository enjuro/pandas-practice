```python
# 3.不要な列、行の削除
## 先頭行と、カラム名（列名）が「〇〇.1」「〇〇.2」となっているものを削除したDataFrame dfを再定義してください。
```

```python
columns_list = df.columns
columns_list
```

```python
### ".1"か".2"で終わるものを除外する（filterをつかう？）
```

```python
target_columns_list = list(filter(lambda column: not(column.endswith(".1") or column.endswith(".2")), columns_list))
target_columns_list
```

```python
df = df[target_columns_list][1:]
df
```

```python
## 答え合わせしたら手動で削除していた。自分はスマートにできてよかった。
```

```python
# 6. 条件抽出
## people.csvを読み込みdf_peopleと定義し
## その後、下記の条件を満たすDataFrameをそれぞれ抽出してください
## ・nationalityがAmericaである
## ・ageが20以上30未満
```

```python
df_people = pd.read_csv("./data/people.csv")
```

```python
df_people
```

```python
df_people["nationality"] == "America"
```

```python
df_people[df_people["nationality"] == "America"] #メソッドを使わない方法
```

```python
df_people.query('nationality == "America"') #引数の条件は文字列
```

```python
df_people[df_people["nationality"].isin(["America"])]
```

```python
df_people["age"] >= 20
```

```python
df_people["age"] < 30
```

```python
df_people[(df_people["age"] >= 20) & (df_people["age"] < 30)]
```

```python
df_people.query('age >= 20 & age < 30')
```

```python
df_people.query('20 <= age < 30') # これでも行ける
```

```python
# 3.不要な列、行の削除
## 先頭行と、カラム名（列名）が「〇〇.1」「〇〇.2」となっているものを削除したDataFrame dfを再定義してください。
```

```python
columns_list = df.columns
columns_list
```

```python
### ".1"か".2"で終わるものを除外する（filterをつかう？）
```

```python
target_columns_list = list(filter(lambda column: not(column.endswith(".1") or column.endswith(".2")), columns_list))
target_columns_list
```

```python
df = df[target_columns_list][1:]
df
```
